```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
img_name = list(uploaded.keys())[0]
img = cv.imread(img_name)

# Convert to RGB for display
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Blur image
blur = cv.GaussianBlur(img, (0, 0), 1.0)
blur_rgb = cv.cvtColor(blur, cv.COLOR_BGR2RGB)

# Unsharp mask sharpening
sharpened = cv.addWeighted(img, 2.0, blur, -1.0, 0)
sharpened_rgb = cv.cvtColor(sharpened, cv.COLOR_BGR2RGB)

# Display results
plt.figure(figsize=(15, 5))

#(a) Original Image
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title('(a) Original Image')
plt.axis('off')

# (b) Blurred Image
plt.subplot(1, 3, 2)
plt.imshow(blur_rgb)
plt.title('(b) Blurred Image')
plt.axis('off')

# (c) Sharpened Image
plt.subplot(1, 3, 3)
plt.imshow(sharpened_rgb)
plt.title('(c) Sharpened Image (Unsharp Mask)')
plt.axis('off')

plt.tight_layout()
plt.show()

```
