```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload salt & pepper image
uploaded = files.upload()
img_name = list(uploaded.keys())[0]
img = cv.imread(img_name)

# (a) Gaussian smoothing
# Uses a 5x5 kernel.
# The noise gets blurry
gauss = cv.GaussianBlur(img, (5, 5), 1.0)

# (b) Median filtering
# Uses a 5x5 kernel.
# The noise is usually completely removed.
median = cv.medianBlur(img, 5)

# Display results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
axs[0].set_title('Original (Salt & Pepper)'); axs[0].axis('off')
axs[1].imshow(cv.cvtColor(gauss, cv.COLOR_BGR2RGB))
axs[1].set_title('Gaussian Blur\n(Noise is blurred, not removed)'); axs[1].axis('off')
axs[2].imshow(cv.cvtColor(median, cv.COLOR_BGR2RGB))
axs[2].set_title('Median Filter\n(Best for this noise)'); axs[2].axis('off')

plt.tight_layout()
plt.show()


```
